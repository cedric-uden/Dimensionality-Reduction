```python
from sklearn.datasets import make_swiss_roll
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use('classic')
```

# Draw the roll

```python
X, t = make_swiss_roll(n_samples=12_500, random_state=42)
swiss_roll_color = plt.cm.nipy_spectral
```

```python
figure_size = 35
fig = plt.figure(figsize=(figure_size+10,figure_size))
ax = fig.add_subplot(222, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=t, cmap=swiss_roll_color)

ax.view_init(10, -65) # rotate the view
ax.set_xlabel("$x_1$", fontsize=18, labelpad=15)
ax.set_ylabel("$x_2$", fontsize=18, labelpad=15)
ax.set_zlabel("$x_3$", fontsize=18, labelpad=10)

plt.show()
```

# Simple projection
